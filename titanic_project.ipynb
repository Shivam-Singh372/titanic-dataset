```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt# pyplot is a submodule and matplotlib is a full library.
import seaborn as sns # Seaborn Enhancce visualization
```

```python
# load titanic data set from seaborn
df=sns.load_dataset('titanic')
#dsiplay 5 rows
df.head()
```

```python
df.isnull().sum() # Checking missing values (True at missing values)
```

```python
df.mode() # Most repeating value in DataFrame
```

```python
# handling missing values pythone
df['age'].fillna(df['age'].median(),inplace=True)
```

```python
df
```

```python
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True) # Fill missing 'embarked_town' with most frequent value
df
```

```python
# Drop 'deck' column(too many missing values)
df.drop(columns=['deck'], inplace = True)
df.isnull().sum() # verify missing value
```

```python
# Check duplicates
print("Duplicates:", df.duplicated().sum())
```

```python
# Removem duplicates
df.drop_duplicates(inplace=True)
print("Duplicate:",df.duplicated().sum())
```

```python
# Boxplot to check outliers in 'age'
plt.figure(figsize=(6,4))
sns.boxplot(x=df['age'])
plt.show()
```

```python
# Remove outliers using IQR method
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3-Q1

# Define limits
lower_limit = Q1-1.5*IQR
upper_limit = Q3-1.5*IQR

# Remove outliers
df=df[(df['age']>= lower_limit) & (df['age']<= upper_limit)]
df
```

```python
# Create a new column 'family_size'
df['Family_size'] = df['sibsp']+ df['parch']+1

#Create a new column 'is_child' (1 if age<16, else 0)
df['is_child'] = np.where(df['age']<16,1,0)
df.head(20)
```

```python
# Countplot of survivors
sns.countplot(x='survived', data=df)
plt.title('Survival count')
plt.show()
```

```python
sns.countplot(x='embarked', data=df)
plt.title('embarked')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=6)
plt.title("Age Distribution")
plt.show()
```

```python
df[df['sex']=='male'].survived.groupby(df.survived).count().plot(kind='pie', figsize=(6, 6), explode=[0, 0.05 ],autopct='%1.1f%%', colors=['red','green'])

plt.legend(["perished","survived"])
plt.title("male survived rate")
plt.show()
```

```python
# Heatmap
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(15,10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='Reds', linewidths=0.5)
plt.title(" FeaturesCorrelation")
plt.show()



```
